```python
# Reload df to ensure it's populated with data and the 'Attendance' column exists.
df = pd.read_csv("/content/sample_data/seminar_attendance.csv")

print("\nStatistical Summary of Attendance:\n")
print(df['Attendance'].describe())
```

```python
plt.figure()
plt.plot(df['Seminar'], df['Attendance'], marker='o')
plt.title("Seminar Attendance Trend")
plt.xlabel("Seminars")
plt.ylabel("Number of Students")
plt.grid(True)
plt.show()
```
