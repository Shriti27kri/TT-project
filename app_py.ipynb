```
import pandas as pd
import matplotlib.pyplot as plt
import ipywidgets as widgets
from IPython.display import display, clear_output

data = {
    "Date": [
        "2025-01-05","2025-01-05","2025-01-05","2025-01-05",
        "2025-01-12","2025-01-12","2025-01-12","2025-01-12",
        "2025-01-19","2025-01-19","2025-01-19","2025-01-19"
    ],
    "Seminar": [
        "AI","AI","AI","AI",
        "ML","ML","ML","ML",
        "DS","DS","DS","DS"
    ],
    "Student_Name": [
        "Rahul Sharma","Priya Singh","Aman Kumar","Neha Verma",
        "Rohit Gupta","Simran Kaur","Vikas Yadav","Pooja Patel",
        "Arjun Mehta","Kunal Shah","Sneha Roy","Anjali Mishra"
    ],
    "Attendance": [
        "Present","Present","Absent","Present",
        "Present","Absent","Present","Present",
        "Present","Present","Absent","Present"
    ]
}
df = pd.DataFrame(data)

print("College Seminar Attendance Analysis")

seminar_dropdown = widgets.Dropdown(
    options=["All"] + list(df["Seminar"].unique()),
    description="Seminar:"
)

analyze_button = widgets.Button(
    description="Analyze Attendance",
    button_style="success"
)

output_area = widgets.Output()

def analyze(b):
    with output_area:
        clear_output()

        selected = seminar_dropdown.value

        if selected == "All":
            filtered_df = df
        else:
            filtered_df = df[df["Seminar"] == selected]

        display(filtered_df)

        plt.figure()
        for seminar in filtered_df["Seminar"].unique():
            s = filtered_df[filtered_df["Seminar"] == seminar]
            plt.plot(s["Date"], s["Attendance"], label=seminar)

        plt.xlabel("Date")
        plt.ylabel("Attendance")
        plt.title("Attendance Trend")
        plt.legend()
        plt.show()

        print("\n Statistics Summary")
        display(filtered_df.groupby("Seminar")["Attendance"].describe())

analyze_button.on_click(analyze)

display(seminar_dropdown, analyze_button, output_area)
```
